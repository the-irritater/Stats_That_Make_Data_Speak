## Setup & Libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from pathlib import Path

# Setup relative root path
ROOT_DIR = Path().resolve().parents[1]
RAW_DATA = ROOT_DIR / "data" / "raw"
print(f"Workspace Root: {ROOT_DIR}")
print(f"Raw Data Folder: {RAW_DATA}")
```

## 1. Verify E-commerce Regression Coefficients

Assert that Session Duration positively predicts spend, and verify standard errors.

```python
df_ecommerce = pd.read_csv(RAW_DATA / "ecommerce.csv")
model = ols('Total_Spend ~ Session_Duration + Pages_Visited', data=df_ecommerce).fit()

# Retrieve parameters
duration_coef = model.params['Session_Duration']
pages_coef = model.params['Pages_Visited']
r_squared = model.rsquared

print("OLS Parameter Verifications:")
print(f"  Session Duration coefficient: {duration_coef:.4f} (Expected ~1.16)")
print(f"  Pages Visited coefficient:    {pages_coef:.4f} (Expected ~3.38)")
print(f"  R-squared:                     {r_squared:.4f} (Expected ~0.71)")

# Tolerance-based assertions for validation
assert abs(duration_coef - 1.16) < 0.1, "Session duration coefficient mismatch"
assert abs(pages_coef - 3.38) < 0.1, "Pages visited coefficient mismatch"
assert abs(r_squared - 0.71) < 0.05, "R-squared mismatch"
print("\n✅ OLS Coefficients Verified Successfully!")
```

## 2. Verify Discount Retention (Chi-Square Test of Independence)

Verify that there is no statistically significant linear relation between discount depth and customer retention.

```python
df_discounts = pd.read_csv(RAW_DATA / "customer_discounts.csv")
contingency = pd.crosstab(df_discounts['Discount_Rate'], df_discounts['Retained'])
chi2, p_val, dof, expected = stats.chi2_contingency(contingency)

print("Chi-Square Test Verifications:")
print(f"  Chi2 Statistic: {chi2:.4f} (Expected ~7.06)")
print(f"  p-value:        {p_val:.4f} (Expected ~0.22)")

assert abs(p_val - 0.22) < 0.05, "Chi-Square p-value mismatch"
print("\n✅ Chi-Square Independence Test Verified Successfully!")
```

## 3. Verify Screen Time Tipping Point (6-Hour Drop-off)

Verify the productivity drop after screen time hours exceed 6.0.

```python
df_screen = pd.read_csv(RAW_DATA / "screen_time.csv")
low_med = df_screen[df_screen['Screen_Time_Hours'] <= 6.0]['Productivity_Score'].mean()
high = df_screen[df_screen['Screen_Time_Hours'] > 6.0]['Productivity_Score'].mean()

print("Productivity Average Verification:")
print(f"  Mean score below 6 hours: {low_med:.2f} (Expected ~67)")
print(f"  Mean score above 6 hours: {high:.2f} (Expected ~46.3)")

assert abs(low_med - 67.0) < 2.0, "Below 6 hours mean mismatch"
assert abs(high - 46.3) < 2.0, "Above 6 hours mean mismatch"
print("\n✅ Screen Time Tipping Point Verified Successfully!")
```
